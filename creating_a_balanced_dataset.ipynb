Reading Training Dataset

```python
def load_and_read_train_data():
  i=0
  train_annotations=[]
  train_images=[]
  for images in os.listdir('/content/train_and_val_set/train_set/images'):
        img_name=images[:-4]+'_'
        img=cv2.imread('/content/train_and_val_set/train_set/images/'+images)
        img=cv2.resize(img,(50,50))
        train_images.append(np.array([img,images]) )
        ann=np.array([np.load('/content/train_and_val_set/train_set/annotations/'+img_name+'aro.npy'),np.load('/content/train_and_val_set/train_set/annotations/'+img_name+'lnd.npy'),np.load('/content/train_and_val_set/train_set/annotations/'+img_name+'val.npy'),np.load('/content/train_and_val_set/train_set/annotations/'+img_name+'exp.npy')])
        train_annotations.append(ann)
        i+=1
        if(i%2000==0): #as there are 200k+ images in training set ive only used 100k images
          print(i)
  return train_annotations,train_images
```

```python
train_annotations,train_images=load_and_read_train_data()
```

```python
#separating images based on labels
train_neutral_images=train_images[np.where((train_annotations[:,3]=='0'))]
train_happy_images=train_images[np.where((train_annotations[:,3]=='1'))]
train_sad_images=train_images[np.where((train_annotations[:,3]=='2'))]
train_surp_images=train_images[np.where((train_annotations[:,3]=='3'))]
train_fear_images=train_images[np.where((train_annotations[:,3]=='4'))]
train_disg_images=train_images[np.where((train_annotations[:,3]=='5'))]
train_anger_images=train_images[np.where((train_annotations[:,3]=='6'))]
train_contempt_images=train_images[np.where((train_annotations[:,3]=='7'))]
```
